```python
import numpy as np
from post_proc import read_config, compile_directory, load_results, animate_imshow
```

```python
# set the diectory with all results in it
direct = "local_random/"
```

```python
# compile the individual csv files at every timestep to a single csv file
compile_directory(direct)
```

```python
# read the confi file to find the iterations to animate for
part_x, part_y, iterations = read_config(direct)

# load the compiled results
results = load_results(direct, "all", 0, iterations)

# setup the amination
gol_ani = animate_imshow(results, dt=1, t0=0)
gol_ani.title = "GOL example"
gol_ani.delay = 20

gol_ani.run(html=True)  # run it
gol_ani.save_html(direct + "_" + "all.html")  # save it
```
